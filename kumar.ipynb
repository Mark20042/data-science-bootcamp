```python
import numpy as np
np.__version__
```

```python
np.array([1, 2, 3])

np.array([5, 6, 8], dtype='float32') # specifying the data type while creating an array by assigning dtype we can also convert the data type of the array

# 2D array
np.array([[1, 2, 3], [4, 5, 6]] , dtype='int32')


# Ways to initialize the array
np.zeros((3,4)) # creating an array of zeros with 3 rows and 4 columns
np.ones((3,4)) # creating an array of ones with 3 rows and 4 columns
np.empty((3,4)) # creating an empty array with 3 rows and 4 columns
np.arange(10, 20, 2) # creating an array with a range of values from 10 to 20 with a step of 2
np.linspace(0, 1, 5) # creating an array with 5 values
# equally spaced between 0 and 1
np.eye(3) # creating an identity matrix of size 3x3

# Array attributes
a = np.array([[1, 2, 3], [4, 5, 6]])
a.ndim # number of dimensions
a.shape # shape of the array
a.dtype # data type of the array
a.size # number of elements in the array
a.itemsize # size of each element in bytes
a.nbytes # total size of the array in bytes
a.reshape(3, 2) # reshaping the array to 3 rows and 2 columns
a.flatten() # flattening the array to 1D
a.T # transpose of the array
a.astype('float32') # converting the data type of the array to float32
a.tolist() # converting the array to a list
a.copy() # creating a copy of the array
a.view() # creating a view of the array
a.sum() # sum of all elements in the array
a.min() # minimum value in the array
a.max() # maximum value in the array
a.mean() # mean of all elements in the array
a.std() # standard deviation of all elements in the array
a.var() # variance of all elements in the array
np.unique(a) # unique elements in the array
np.sort(a) # sorting the array
np.concatenate((a, a), axis=0) # concatenating two arrays along rows
np.concatenate((a, a), axis=1) # concatenating two arrays along columns
np.vstack((a, a)) # stacking two arrays vertically
np.hstack((a, a)) # stacking two arrays horizontally
np.split(a, 2) # splitting the array into two equal parts
np.where(a > 3) # indices of elements greater than 3
np.extract(a > 3, a) # extracting elements greater than 3
np.nonzero(a) # indices of non-zero elements
np.save('array.npy', a) # saving the array to a file
np.load('array.npy') # loading the array from a file
# np.savez('array.npz', a=a, b=b) # saving multiple arrays to a file
# np.load('array.npz') # loading multiple arrays from a file
# np.savetxt('array.txt', a) # saving the array to a text file
# np.loadtxt('array.txt') # loading the array from a text file
# np.genfromtxt('array.txt') # loading the array from a text file with missing values
# np.savetxt('array.csv', a, delimiter=',') # saving the array to a CSV file
# np.genfromtxt('array.csv', delimiter=',') # loading the array from a CSV file
np.random.rand(3, 4) # creating an array with random values between 0 and 1
np.random.randint(0, 10, (3, 4)) # creating an array with random integers between 0 and 10
np.random.normal(0, 1, (3, 4)) # creating an array with random values from a normal distribution with mean 0 and standard deviation 1
np.random.uniform(0, 1, (3, 4)) # creating an array with random values between 0 and 1
np.random.seed(42) # setting the seed for random number generation
np.random.rand(3, 4) # creating an array with random values between 0 and 1

```
